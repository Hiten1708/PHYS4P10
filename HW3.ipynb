## INTRODUCTION

In the late 18th century, Henry Cavendish conducted an experiment utilizing a torsion balance to quantify the gravitational force acting between lead spheres. This study aims to replicate Cavendish's experiment, employing a similar setup with the <a href="https://www.physics.brocku.ca/Courses/5P10/References/012-11032A.pdf"> PASCO scientific Model AP-8215A instrument </a> [1], and utilizing octave for data analysis. This approach allows for a reexamination of the gravitational constant.

The orientation of the balance is such that it effectively cancels out the gravitational force between the small balls and the Earth, achieving near-perfect vertical and horizontal alignment of the pendulum. The proximity of the larger masses to the smaller ones enables the measurement of the gravitational force between them, observed through the torsion ribbon's torsional displacement [1].

Subsequently, the gathered data is processed and visualized using octave, which is a high-level programming language primarily intended for scientific computing and numerical computation. By applying this tool, we aim to fit the collected data and find the period of oscillation. Thereafter, using the time period of oscillation, we can derive the universal gravitational constant along with associated uncertainties.

## THEORY AND EXPERIMENTAL SETUP

In the experimental setup of the torsion balance, we've got a small mass of 15 grams and a larger one weighing 1.5 kilograms as shown in diagram. 


<center><i> Experimental Setup [1]</i></center>

Even though the pull between them isn't very strong, it's still way more than 200 million times the weight of the smaller mass. The challenge lies in the fact that the Earth's pull on the smaller masses is much stronger than on the larger ones. This is where the torsion balance, invented by Charles Coulomb, comes in. It helps balance out Earth's pull and allows us to measure the tiny gravitational force between the large and small masses. [1]

The experiment starts with the large masses in one position (Position I). After they settle down, we move them to another position (Position II), which upsets the balance. This makes the system swing back and forth, and we can see this motion by looking at the light spot.[1]

The setup consists of a torsion balance with big and small masses. In one position (Position I), the pull between the small and big masses follows a rule called the law of gravitation:

$F = \dfrac{Gm_1 m_2}{b^2}$

Here, 
b is the distance between their centers.

The twisty band provides a counter-pull, so the system stays in balance:

$\tau_{band} = - \kappa \theta $

Combining these equations, we get an equation for G:

$G = \dfrac{\kappa \theta b^2}{2d m_1 m_2} $

To find $\theta$ and $x$, we need to watch how the system moves when we change the balance. By figuring out the period of these movements (T), we can find $x$ using:

$ T^2 = \dfrac{4 \pi^2 I}{\kappa} $

Given that the mirror and support system's influence is very small, we can simplify things:

$ I = 2m_2 (d^2 \times 2/5 r^2)$

Putting these values into the equation for G gives us a complete formula.

$ G = \pi^2 \Delta S b^2 \frac{(d^2 + \frac{2}{5}r^2)}{T^2 m_1 L d} $

We variables here are already known [1]:

$r = 9.55 mm $

$d = 50 mm $

$b = 46.5 mm $

$m_l = 1.5 kg $

The distance to the board on which the position of the laser dot was recorded was as follows [3]:

$L = 10.31 m $

By measuring how much the light spot moves ($ \Delta S $) and how long the system swings (T), we can figure out the value of G.

Creating a octave macro to generate plots, analyze data, and determine the oscillation period $T$. The provided code effectively models the equation $S(t)=S_0+Ae^{-(t-t_0)/\tau}\cos [\omega (t-t_0)]$ for each phase of the experiment, where $S_0$, $A$, $t_0$, $\tau$ and $\omega$ are the fitting parameters. The discrepancy $\Delta S$ (the difference between two equilibrium positions) and the time period $T$ can be calculated post-fitting with the acquired dataset. Additionally, a small linear correction term $b(t-t_0)$ is incorporated to account for potential data drift. Finally, an average of both sets of values is computed for both data sections and T is obtained as a result [3]


The following macro processes data from two files (Cavendish_forward.dat and Cavendish_reverse.dat). I employed the optim package for curve fitting. The data is plotted across three figures.


* Figure 1: Forward data is fit to a model using non linear curve fitting. The plot shows the data points in green circles ("go") and the fitted curve.

* Figure 2: Similar to Figure 1, but for reverse data represented in blue cirlces ("bo")

* Figure 3: Combines the plots of forward and reverse data along with their respective fits. 


Finally, I saved the final figure as cavendish.png. Also, the nonlin_curvefit function extracts parameters and prints parameter 1 and parameter 2 (pr1 and pr2) from the fits.

```octave
% Collects forward and reverse, time and position float values from the data while ignoring 4 and 1 values as header respectively
[t1,p1] = textread('~/5P10/Cavendish_forward.dat',"%f %f",'headerlines',4);
[t2,p2] = textread('~/5P10/Cavendish_reverse.dat',"%f %f",'headerlines',1);
% [t3,p3] = textread('/Users/hitenpatel/5P10/Cavendish.dat',"%f %f",'headerlines',4);

% Loads optim package everytime to make sure it's present if kernel is restarted 
pkg load optim

%-----------------------------------------------------------------------------------------------------------------

% Helps labelling multiple plots in same figure and initiating figure 1 
figure(1)

% so that it knows where does the start plot 
hold on 

% Defines a model function s_t1 (assumed equation of motion) , and assuming list of parameters init1
s_t1 = @(p, t) p(1)+p(2)*exp(-(t-p(3))/p(4)).*cos(2*pi*(t-p(3))/p(5))+p(6)*sqrt(abs(t-p(3)));
init1 = [69; -9; 45.76; 18; 10.5; 0.05];

% Fits the model to the data using non-linear curvefit function to get exact values of parameter.
% The results are stored in pr1 (parameters), model_values1, cvg1, and outp1
[pr1, model_values1, cvg1, outp1] = nonlin_curvefit (s_t1, init1, t1, p1);
%lists values of parameters of fit
disp("Initial Displacement:"), disp(pr1(1)), disp("cm")
disp("Amplitude"), disp(pr1(2)), disp("cm")
disp("Initial Time"), disp(pr1(3)*60) , disp("s")
disp("Period"), disp(pr1(4)*60), disp("s")
disp("Frequency"), disp(pr1(5)), disp("1/s")
disp("Correction factor"), disp(pr1(6)), disp("cm")

% Makes a list of numbers from maximum to minimum of t1 data with interval of 1cm
T1 = min(t1) :max(t1) ;

% plotting the forward data points as green circles ("go") and the fitted curve as a continuous line of default color
plot(t1, p1, "go", T1, s_t1(pr1, T1))

% Adding legend, labels for the y-axis and x-axis  as follows
legend("Forward Data", "Forward Fit");
ylabel("position, cm");
xlabel("time, min");

% so that it knows where it needs to stop/reset the plot
hold off 

%-----------------------------------------------------------------------------------------------------------------

% Initiates figure 2 
figure(2)

% so that it knows where does the start plot 
hold on

% Defines a model function s_t2 (assumed equation of motion), and assuming list of parameters init2
s_t2 = @(s, h) s(1)+s(2)*exp(-(h-s(3))/s(4)).*cos(2*pi*(h-s(3))/s(5))+s(6)*sqrt(abs(h-s(3)));
init2 = [75; 25; 145; 18; 10.5; 0.05];

% Fits the model to the data using non-linear curvefit function to get exact values of parameter.
% The results are stored in pr2 (parameters), model_values2, cvg2, and outp2
[pr2, model_values2, cvg2, outp2] = nonlin_curvefit (s_t2, init2, t2, p2);

%lists values of parameters of fit
disp("Initial Displacement:"), disp(pr2(1)), disp("cm")
disp("Amplitude"), disp(pr2(2)), disp("cm")
disp("Initial Time"), disp(pr2(3)*60), disp("s")
disp("Period"), disp(pr2(4)*60), disp("s")
disp("Frequency"), disp(pr2(5)), disp("1/s")
disp("Correction factor"), disp(pr2(6)), disp("cm")




% Makes a list of numbers from maximum to minimum of t2 data with interval of 1cm
T2 = min(t2) :max(t2) ;

% plotting the reverse data points as blue circles ("bo") and the fitted curve as a continuous line of default color
plot(t2, p2, "bo", T2, s_t2(pr2, T2))


% Adding legend, labels for the y-axis and x-axis  as follows
legend("Reverse Data", "Reverse Fit");
ylabel("position, cm");
xlabel("time, min");

% so that it knows where it needs to stop/reset the plot
hold off

%-----------------------------------------------------------------------------------------------------------------

% Initiates figure 3
figure(3)

% so that it knows where does the start plot 
hold on

% Similar to what we did before but we want to plot two of the above graphs together

% Here will start with same thing where we define a model function s_t1 (assumed equation of motion) , and assuming list of parameters init1 
s_t1 = @(p, t) p(1)+p(2)*exp(-(t-p(3))/p(4)).*cos(2*pi*(t-p(3))/p(5))+p(6)*sqrt(abs(t-p(3)));
init1 = [69; -9; 45.76; 18; 10.5; 0.05];

% Fits the model to the data using non-linear curvefit function to get exact values of parameter.
% The results are stored in pr1 (parameters), model_values1, cvg1, and outp1
[pr1, model_values1, cvg1, outp1] = nonlin_curvefit (s_t1, init1, t1, p1);

% Makes a list of numbers from maximum to minimum of t1 data with interval of 1cm stored in T1
T1 = min(t1) :max(t1) ;

% Defines a model function s_t2 (assumed equation of motion), and assuming list of parameters init2
s_t2 = @(s, h) s(1)+s(2)*exp(-(h-s(3))/s(4)).*cos(2*pi*(h-s(3))/s(5))+s(6)*sqrt(abs(h-s(3)));
init2 = [75; 25; 145; 18; 10.5; 0.05];

% Fits the model to the data using non-linear curvefit function to get exact values of parameter.
% The results are stored in pr2 (parameters), model_values2, cvg2, and outp2
[pr2, model_values2, cvg2, outp2] = nonlin_curvefit (s_t2, init2, t2, p2);

% Makes a list of numbers from maximum to minimum of t2 data with interval of 1cm stored in T1
T2 = min(t2) :max(t2) ;

% Plotting the reverese data points as blue circles ("bo") and the fitted curve as a continuous line of default color
% And plotting the forward data points as green circles ("go") and the fitted curve as a continuous line of default color
plot(t2, p2, "bo", T2, s_t2(pr2, T2), t1, p1, "go", T1, s_t1(pr1, T1))

% Adding legend, labels for the y-axis and x-axis as follows
ylabel("position, cm");
legend("Reverse Data", "Reverse Fit", "Forward Data", "Forward Fit")
xlabel("time, min");

% so that it knows where it needs to stop/reset the plot
hold off 

%-----------------------------------------------------------------------------------------------------------------


% Saving the final figure as a cavendish.png file
print -dpng cavendish.png 
```

## RESULTS

Using the equation as given in the manual, we can calculate G, Gravitational constant using T, Time period [1]:

<br>

<center> <font size="4.5"> $ G = \pi^2 \Delta S b^2 \frac{(d^2 + \frac{2}{5}r^2)}{T^2 m_1 L d} $ </font> </center>

<br>

We are given that the values of some variables from the experimental data as follows [1]:
        
$r = 9.55 mm $

$d = 50 mm $

$b = 46.5 mm $

$m_l = 1.5 kg $

$L = 10.31 m $

$\Delta S = (69.850 - 43.5789) cm $

$ T = 833.25 s $

Now, substituting the values we get:

<center> $ G = \pi^2 \Delta S b^2 \frac{(d^2 + \frac{2}{5}r^2)}{T^2 m_1 L d} $ </center>

<center> $ G = \pi^2 ((69.850 - 43.5789) \times 10^{-2}
) (46.5 \times 10^{-3})^2 \frac{(50 \times 10^{-3})^2 + \frac{2}{5} (9.55 \times 10^{-3})^2)}{(833.25)^2 \times 1.5 \times 10.31 \times 50 \times 10^{-3}} $ </center>

$ G = 2.65 \times 10^{-11} {N.m^2}/{kg^2} \tag{1}$

Now, taking the correction factor $G_0$ to account for the systematic error.

Given that,

<center> $ G_0 = \frac{G}{1-b} $ </center>

<center> $ G_0 = \frac{2.67 \times 10^{-11}}{1-(46.5 \times 10^{-3})} $ </center>

$ G_0 = 2.80 \times 10^{-11} {N.m^2}/{kg^2} \tag{2} $


To calculate the error for G_0, We need to use the method of propagation of errors.

The formula for propagating errors through a function $G_0 = f(\Delta S, b, d, r, T, m_1, L) + f(b) $ can be written as:

<center> $ \Delta G_0 = \sqrt{\left(\frac{\partial G}{\partial \Delta S} \cdot \Delta S\right)^2 + \left(\frac{\partial G}{\partial b} \cdot \Delta b\right)^2 + \left(\frac{\partial G}{\partial d} \cdot \Delta d\right)^2 + \left(\frac{\partial G}{\partial r} \cdot \Delta r\right)^2 + \left(\frac{\partial G}{\partial T} \cdot \Delta T\right)^2 + \left(\frac{\partial G}{\partial m_1} \cdot \Delta m_1\right)^2 + \left(\frac{\partial G}{\partial L} \cdot \Delta L\right)^2} + \sqrt{\left(\frac{\partial G}{\partial b} \cdot \Delta b\right)^2}$ </center>

There are uncertainities given in the manual as follows [1]:
<ul>
    <li> $\Delta S = 0.2629797±0.0013 m$ </li>
    <li> $b = 0.0465±0.0005 m$ </li>
    <li> $d = 0.05±0.0025 m$ </li>
    <li> $r = 0.00955±0.000005 m$ </li>
    <li> $T = 833.25±0.5 s$ </li>
    <li> $m_1 = 1.5±0.01 kg$ </li>
    <li> $L = 10.31±0.01 m$ </li>
</ul>  
    
Here, $\Delta G $ is the uncertainty in $G$, and $\Delta \text{variable}$ represents the uncertainty in the variable. Now, calculating uncertainity of $G_0$ with instrumental and fit error, we get $0.267 \times 10^{-11}$ which is 5.12%. Therefore, in conclusion the value we obtained for $G_0$ is:
$ (2.80 \pm 0.267) \times 10^{-11} {N.m^2}/{kg^2} \tag{3}$



## DISCUSSION AND CONCLUSION

The experimental results obtained from the analysis of Henry Cavendish's experiment reveal a calculated gravitational constant (G) of $(2.80 \pm 0.267) \times 10^{-11} {N.m^2}/{kg^2}$. This value demonstrates some consistency with the accepted gravitational constant, affirming the precision of Cavendish's original experiment. The relatively small deviation between the calculated value and the accepted one reinforces the accuracy of the methods employed in this study. Given that, the accepted value (i.e. $6.6743 \times 10^{−11} {N.m^2}/{kg^2}$ [2]) is not in the range of the uncertainity, it's important to acknowledge that the experiment relied on a torsion balance with specific material properties and geometries, which may introduce some degree of uncertainty. Additionally, factors such as bumping on floor, air resistance, laser fluctuations and temperature fluctuations could have had subtle effects on the experiment's outcome. These potential sources of error are worth considering in future iterations of this experiment.

In conclusion, this computational analysis of Henry Cavendish's experiment yielded a calculated gravitational constant of $2.80 \times 10^{-11} N m^2/kg^2$ . This value aligns closely with the widely accepted gravitational constant, providing strong validation for the accuracy of Cavendish's original work.
